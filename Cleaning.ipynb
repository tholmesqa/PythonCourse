## Missing Values

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'particpant': [1, 2, 3, 4],
    'age': [50, None, 30, np.nan],
    'satisfaction': [None, 8, 9, None]
})
```

```python
df
```

```python
df.isna().any()
```

```python
df.isna().sum()
```

```python
df.isna().any(axis=1)
```

```python
df[df.isna().any(axis=1)]
```

```python
df.dropna()
```

```python
df.dropna(thresh=2)
```

```python
df.dropna(subset=['age'])
```

```python
df.fillna({'age': df['age'].mean()})
```

```python
df.fillna(10, inplace=True)
```

## Transformations

```python
df = pd.read_csv('data/loan_data.csv')
```

```python
df.drop('ID', axis=1, inplace=True)
```

```python
df.drop(0)
```

```python
df.replace({
    'Long Term': '12 Month',
    'Short Term': '6 Month'
})
```

```python
df['Debt']
```

```python
def debtbin(debt):
    if debt > 1000:
        return 'High'
    else:
        return 'Low'
```

```python
debtbin(1002)
```

```python
df['Debt'].map(debtbin)
```

```python
df['Debt'].map(lambda debt: 'High' if debt > 1000 else 'Low')
```

```python
num_df = df.select_dtypes('number')
```

```python
num_df.apply(lambda col: col.round(2))
```
