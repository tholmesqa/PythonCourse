# Visualisation

```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/loan_data.csv')
```

## Matplotlib

```python
fig, ax = plt.subplots(figsize=(8, 4))

# MATLAB style
# plt.scatter(df['Debt'], df['Balance'])
# plt.title('Debt vs. Balance')
# plt.xlabel('Debt')
# plt.ylabel('Balance')

# OO style
ax.scatter(df['Debt'], df['Balance'])
ax.set_title('Debt vs. Balance')
ax.set_xlabel('Debt')
ax.set_ylabel('Balance')
```

## Seaborn

```python
sns.scatterplot(data=df, x='Debt', y='Balance', hue='Term')
plt.title('Debt vs Balance');
```

```python
sns.histplot(data=df, x='Debt', hue='Term')
```

```python
sns.boxplot(data=df, x='Default', y='Debt')
```

```python
df_corr = df.corr(numeric_only=True)
```

```python
df_corr
```

```python
sns.heatmap(df_corr,
           center=0.0,
           cmap='coolwarm',
           vmin=-1)
```

```python
sns.pairplot(df.select_dtypes('number'))
plt.savefig('scattermatrix.png')
```
